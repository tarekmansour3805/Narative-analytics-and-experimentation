```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
Craches= pd.read_csv('plane_crashes_data.csv')
```

```python
Craches.head()
```

```python
Craches.year.unique()
```

```python
Craches.dropna(subset=["year"], inplace=True)
Craches.dropna(subset=["month"], inplace=True)
```

```python
craches2 = Craches.loc[(Craches['year'] >= 1950) & (Craches['year'] < 1980)]
craches3 = Craches.loc[(Craches['year'] >= 1980) & (Craches['year'] <= 2009)]
```

```python
print('1950s Fatalities size:', len(craches2.fatalities))
print('1980s Fatalities size:', len(craches3.fatalities))
print("\n")
print('1950s Sum of fatlities:', craches2.fatalities.sum())
print('1980s Sum of fatlities:', craches3.fatalities.sum())
print("\n")
print('1950s Mean:', craches2.fatalities.mean())
print('1080s Mean:', craches3.fatalities.mean())
print("\n")
print('Sum of all Fatalities:', Craches.fatalities.sum())
print('Mean of all Fatalities:', Craches.fatalities.mean())
print('STD of all Fatalities:', Craches.fatalities.std())
print("\n")
print(stats.ttest_ind((craches2).fatalities,(craches3).fatalities))
print()


```

```python
print('from june to December')
plt.plot(craches2.loc[1950:1980]['fatalities'])
plt.show()

print('from june to December')
plt.plot(craches3.loc[1980:2009]['fatalities'])
plt.show()
```

```python
fig = craches2[
    ['month', 'fatalities']].groupby('month').count().plot()
craches3[
    ['month', 'fatalities']].groupby('month').count().plot(ax=fig)
plt.legend(['1950s','1980s'])
plt.title('Comparing the count of samples')
plt.show()

fig = craches2[
    ['month', 'fatalities']].groupby('month').sum().plot()
craches3[
    ['month', 'fatalities']].groupby('month').sum().plot(ax=fig)
plt.legend(['1950s','1980s'])
plt.title('Comparing the sum of samples')
plt.show()

fig = craches2[
    ['month', 'fatalities']].groupby('month').mean().plot()
craches3[
    ['month', 'fatalities']].groupby('month').mean().plot(ax=fig)
plt.legend(['1950s','1980s'])
plt.title('Comparing the mean of samples')
plt.show()


fig = craches2[
    ['month', 'fatalities']].groupby('month').std().plot()
craches3[
    ['month', 'fatalities']].groupby('month').std().plot(ax=fig)
plt.legend(['1950s','1980s'])
plt.title('Comparing the STD of samples')
plt.show()
```
